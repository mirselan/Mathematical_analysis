#Тема “Последовательность”

***1. Как относятся друг к другу множество и последовательность? (в ответе использовать слова типа: часть, целое, общее, частное, родитель, дочерний субъект и т.д.)***

Множеством является совокупность произвольного числа элементов.   

Последовательность - это упорядоченное множество элементов, причём порядок определяется тем, что у каждого элемента последовательности есть порядковый номер, который постоянно возрастает. Также каждый член последовательности имеет свой однозначно определённый номер, т.е. место. 

Например, числовая последовательность - это упорядоченное множество вещественных чисел.  

Любое множество можно вполне упорядочить - это гласит одна из теорем теории множеств (теорема Цермело).

$\Bigl\{a_n\Bigr\}_{n=1}^\infty=2^n - n$

**a)**

Рассмотрим два соседних члена последовательности: $a_n=2^n-n$ и $a_{n+1}=2^{n+1}-n+1$ . Нам нужно выяснить, можно ли утверждать, что при любом $n$  $a_n$ обязательно меньше или больше, чем $a_{n+1}$, то есть, нам нужно сравнить эти два числа. Для этого вычтем из $a_{n+1}$ $a_n$
и посмотрим на знак разности:

$a_{n+1}-a_n=(2^{n+1}-n+1)-(2^n-n)=2^{n+1}+1-2^n$

Очевидно, что последнее выражение положительно при любом значении $n$ от 1 до $\infty$, значит, $a_{n+1}-a_n>0$, что по определению последовательность $\Bigl\{a_n\Bigr\}_{n=1}^\infty$ возрастает, т.е. является монотонной.

**б)**

Мы доказали, что $a_{n+1}-a_n>0$. Значит каждый следующий член последовательности больше предыдущего, т.е. последовательность ограничена снизу.

**в)**

$\Bigl\{a_n\Bigr\}_{n=1}^\infty=2^n - n$

$a_1=2^1-1=1$

$a_2=2^{n+1}-n+1=2^{1+1}-1+1=4-2=2$

$a_5=2^{n+4}-n+4=2^5-5=32$

$\Bigl\{b_n\Bigr\}_{n=2}^\infty=\frac{1}{1 - n}$

**a)**

Сравним между собой числа $b_n=\frac{1}{1-n}$ и $b_{n+1}=\frac{1}{1-(n+1)}$.

$b_{n+1}-b_n=\frac{1}{1-(n+1)}-\frac{1}{1-n}=\frac{(1-n)-(1-(n+1))}{(1-(n+1))*(1-n)}=\frac{1-n-1-n-1}{(1-n-1)*(1-n)}=\frac{-2n-1}{-n(1-n)}=\frac{-n(2+\frac{1}{n})}{-n(1-n)}=\frac{2+\frac{1}{n}}{1-n}$ 

Очевидно, что последнее выражение отрицательно при любом значении $n$ от 2 до $\infty$, значит, $b_{n+1}-b_n<0$, и что по определению последовательность $\Bigl\{b_n\Bigr\}_{n=2}^\infty$ убывает, т.е. является монотонной.

**б)**

Можно заметить, что, уже доказав монотонность этой последовательности, легко доказать ее ограниченность. В самом деле, раз каждый следующий член этой последовательности меньше предыдущего, значит, первый член последовательности - самый большой из всех, что в ней встречаются. Значит, эта последовательность ограничена сверху.

**в)**

$\Bigl\{b_n\Bigr\}_{n=2}^\infty=\frac{1}{1 - n}$

$b_1=\frac{1}{1-n}=\frac{1}{1-2}=-1$

$b_2=\frac{1}{1-(n+1)}=\frac{1}{1-(2+1)}=-\frac{1}{2}$

$b_5=\frac{1}{1-(n+4)}=\frac{1}{1-(2+4)}=-\frac{1}{5}$

$\Bigl\{c_n\Bigr\}_{n=1}^\infty=-1^n+\sqrt[]{2n}$

**a)**

Рассмотрим два соседних члена последовательности: $c_n=-1^n+\sqrt{2n}$ и $c_{n+1}=-1^{n+1}+\sqrt{2(n+1)}$ . Нам нужно выяснить, можно ли утверждать, что при любом $n$  $c_n$ обязательно меньше или больше, чем $c_{n+1}$, то есть, нам нужно сравнить эти два числа. Для этого вычтем из $c_{n+1}$ $c_n$
и посмотрим на знак разности:

$c_{n+1}-c_n=(-1^{n+1}+\sqrt{2(n+1)})-(-1^n+\sqrt{2n})=-1^{n+1}+\sqrt{2n+2}+1^n-\sqrt{2n})$

Очевидно, что последнее выражение положительно при любом значении $n$ от 1 до $\infty$, значит, $c_{n+1}-c_n>0$, и что по определению последовательность $\Bigl\{с_n\Bigr\}_{n=1}^\infty$ возрастает, т.е. является монотонной.

**б)**

Мы доказали, что $с_{n+1}-с_n>0$. Значит каждый следующий член последовательности больше предыдущего, т.е. последовательность ограничена снизу.

**в)**

$\Bigl\{c_n\Bigr\}_{n=1}^\infty=-1^n+\sqrt[]{2n}$

$с_1=-1^n+\sqrt[]{2n}=-1^1+\sqrt[]{2*1}=-1+\sqrt[]{2}$

$с_2=-1^{n+1}+\sqrt[]{2(n+1)}=-1^2+\sqrt[]{2*2}=1+\sqrt[]{4}=3$

$с_5=-1^{n+4}+\sqrt[]{2(n+4)}=-1^5+\sqrt[]{2*5}=-1+\sqrt[]{10}$









$\Bigl\{d_n\Bigr\}_{n=1}^\infty=(-1)^{2n}+\frac{1}{n^2}$

**a)**

Сравним между собой числа $d_n=(-1)^{2n}+\frac{1}{n^2}$ и $d_{n+1}=(-1)^{2(n+1)}+\frac{1}{(n+1)^2}$.

$d_{n+1}-d_n=((-1)^{2(n+1)}+\frac{1}{(n+1)^2})-((-1)^{2n}+\frac{1}{n^2})=((-1)^{2n+2}+\frac{1}{n^2+2n+1})-((-1)^{2n}+\frac{1}{n^2})=(-1)^{2n+2}+\frac{1}{n^2+2n+1}+1^{2n}-\frac{1}{n^2}=$ 

Очевидно, что последнее выражение положительно при любом значении $n$ от 1 до $\infty$, значит, $d_{n+1}-d_n>0$, и что по определению последовательность $\Bigl\{d_n\Bigr\}_{n=1}^\infty$ возрастает, т.е. является монотонной.

**б)**

Можно заметить, что, уже доказав монотонность этой последовательности, легко доказать ее ограниченность. В самом деле, раз каждый следующий член этой последовательности больше предыдущего, значит, первый член последовательности - самый наименьший из всех, что в ней встречаются. Значит, эта последовательность ограничена снизу.

**в)**

$\Bigl\{d_n\Bigr\}_{n=1}^\infty=(-1)^{2n}+\frac{1}{n^2}$

$d_1=(-1)^{2n}+\frac{1}{n^2}=(-1)^{2*1}+\frac{1}{1^2}=1+1=2$

$d_2=(-1)^{2(n+1)}+\frac{1}{(n+1)^2}=(-1)^{2*2}+\frac{1}{2^2}=1+\frac{1}{4}=1\frac{1}{4}$

$d_5=(-1)^{2(n+4)}+\frac{1}{(n+4)^2}=(-1)^{2*5}+\frac{1}{5^2}=1+\frac{1}{25}=1\frac{1}{25}$
