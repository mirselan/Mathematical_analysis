#$U=3-8x+6y$, если $x^2+y^2=36$

Составим функцию Лагранжа:

$L(\lambda,x,y)=3-8x+6y+\lambda\cdot(x^2+y^2-36)$

Найдем частные производные, приравняем их к нулю и решим систему уравнений:

$$\begin{cases}
   L^{\prime}_x=-8+2x\lambda=0\\
   L^{\prime}_y=6+2y\lambda=0\\
   L^{\prime}_{\lambda}=x^2+y^2-36=0
 \end{cases}$$
 
Выразим из первого уравнения $x$, из второго $y$ и подставим всё в третье:

 $$\begin{cases}
   x=\frac{4}{\lambda},\\
   y=-\frac{3}{\lambda},\\
   \Bigr(\frac{4}{\lambda}\Bigl)^2+\Bigr(-\frac{3}{\lambda}\Bigl)^2=36
 \end{cases}
 \Rightarrow
 \begin{cases}
   x=\frac{4}{\lambda},\\
   y=-\frac{3}{\lambda},\\
   \frac{25}{\lambda^2}=36
 \end{cases}
 \Rightarrow
 \begin{cases}
   x=\frac{4}{\lambda},\\
   y=-\frac{3}{\lambda},\\
   \lambda_1^2=\frac{25}{36}
 \end{cases}
 \Rightarrow
 \begin{cases}
   x=\frac{4}{\lambda1},\\
   y=-\frac{3}{\lambda},\\
   \lambda_1=\frac{5}{6}
 \end{cases}$$

 Получаем две точки: $(\frac{24}{5},-\frac{18}{5},\frac{5}{6})$ и $(-\frac{24}{5},\frac{18}{5},-\frac{5}{6})$

Найдем вторые производные:

 $$L^{\prime}_x=-8+2x\lambda\\
   L^{\prime}_y=6+2y\lambda\\
   L^{\prime}_{\lambda}=x^2+y^2-36$$

   $$L^{\prime\prime}_{xx}=2\lambda, \ \ \ L^{\prime\prime}_{yy}=2\lambda, \ \ \ L^{\prime\prime}_{\lambda\lambda}=0$$

   $$L^{\prime\prime}_{xy}=0, \ \ \ L^{\prime\prime}_{x\lambda}=2x, \ \ \ L^{\prime\prime}_{y\lambda}=2y$$

Составим матрицу Гёссе:

   $$\begin{pmatrix}
L''_{\lambda \lambda} & L''_{\lambda x} & L''_{\lambda y}\\ 
L''_{x \lambda} & L''_{xx} & L''_{xy}\\
L''_{y \lambda} & L''_{yx} & L''_{yy}\\
\end{pmatrix}
=\begin{pmatrix}
0 & 2x & 2y\\ 
2x & 2\lambda & 0\\
2y & 0 & 2\lambda\\
\end{pmatrix}$$

Далее найдем определитель в общем виде и разберемся с каждой точкой:

$$\Delta=\begin{vmatrix}
0 & 2x & 2y\\ 
2x & 2\lambda & 0\\
2y & 0 & 2\lambda\\
\end{vmatrix}=
\begin{vmatrix}
+ & - & +\\ 
- & + & -\\
+ & - & +\\
\end{vmatrix}=0\cdot
\begin{vmatrix}
2\lambda & 0\\
0 & 2\lambda\\
\end{vmatrix}-2x\cdot
\begin{vmatrix}
2x & 0\\
2y & 2\lambda\\
\end{vmatrix}+2y\cdot
\begin{vmatrix}
2x & 2\lambda\\
2y & 0\\
\end{vmatrix}=(-2x)\cdot 4x\lambda+2y \cdot(-4y\lambda)=-8\lambda(x^2+y^2)$$

По условию задачи $x^2+y^2=36$, тогда 
$$\Delta=-8\lambda \cdot 36 = -288\lambda$$

Т.е. знак определителя зависит только от знака $\lambda$.

Если $\lambda=\frac{5}{6}$, то $\Delta<0$, следовательно $(\frac{24}{5},-\frac{18}{5},\frac{5}{6})$ - точка минимума.

Если $\lambda=-\frac{5}{6}$, то $\Delta>0$, следовательно $(-\frac{24}{5},\frac{18}{5},-\frac{5}{6})$ - точка максимума.

#$U=2x^2+12xy+32y^2+15$, если $x^2+16y^2=64$

Составим функцию Лагранжа:

$L(\lambda,x,y)=2x^2+12xy+32y^2+15+\lambda\cdot(x^2+16y^2-64)$

Найдем частные производные, приравняем их к нулю и решим систему уравнений:

$$\begin{cases}
   L^{\prime}_x=4x+12y+2x\lambda=0\\
   L^{\prime}_y=12x+64y+32y\lambda=0\\
   L^{\prime}_{\lambda}=x^2+16y^2-64=0
 \end{cases}$$
 
Выразим из третьего уравнения $x$ и $y$ и подставим в первое и второе уравнения соответственно, затем получившиеся в первом и втором уравнениях значения $x$ и $y$ подставляем в третье уравнение и находим $\lambda$:

 $$\begin{cases}
   4x+2x\lambda=-12y,\\
   64y+32y\lambda=-12x,\\
   x^2+16y^2=64
 \end{cases}
 \Rightarrow
 \begin{cases}
   4x+2x\lambda=-12y,\\
   64y+32y\lambda=-12x,\\
   x=8-4y;4y=8-x
 \end{cases}
 \Rightarrow
 \begin{cases}
   4x+2x\lambda=\frac{-12(8-x)}{4},\\
   64y+32y\lambda=-12(8-4y),\\
   x=8-4y;y=\frac{8-x}{4}
 \end{cases}
 \Rightarrow
 \begin{cases}
   4x+2x\lambda=-24+3x,\\
   64y+32y\lambda=-96+48y,\\
   x^2+16y^2=64
 \end{cases}
 \Rightarrow
 \begin{cases}
   4x-3x+2x\lambda=-24,\\
   64y-48y+32y\lambda=-96,\\
   x^2+16y^2=64
 \end{cases}
 \Rightarrow
 \begin{cases}
   x+2x\lambda=-24,\\
   16y+32y\lambda=-96,\\
   x^2+16y^2=64
 \end{cases}
 \Rightarrow
 \begin{cases}
   x(1+2\lambda)=-24,\\
   16y(1+2\lambda)=-96,\\
   x^2+16y^2=64
 \end{cases}
 \Rightarrow
 \begin{cases}
   x=\frac{-24}{1+2\lambda},\\
   16y=\frac{-96}{1+2\lambda},\\
   x^2+16y^2=64
 \end{cases}
 \Rightarrow
 \begin{cases}
   x=\frac{-24}{1+2\lambda},\\
   y=\frac{-96}{16(1+2\lambda)},\\
   x^2+16y^2=64
 \end{cases}
 \Rightarrow
 \begin{cases}
   x=\frac{-24}{1+2\lambda},\\
   y=\frac{-6}{1+2\lambda},\\
   \Bigl(\frac{-24}{1+2\lambda}\Bigr)^2+16\Bigl(\frac{-6}{1+2\lambda}\Bigr)^2=64
 \end{cases}
 \Rightarrow
 \begin{cases}
   x=\frac{-24}{1+2\lambda},\\
   y=\frac{-6}{1+2\lambda},\\
   (\frac{576}{(1+2\lambda)^2}+\frac{576}{(1+2\lambda)^2}=64
 \end{cases}
 \Rightarrow
 \begin{cases}
   x=\frac{-24}{1+2\lambda},\\
   y=\frac{-6}{1+2\lambda},\\
   (1+2\lambda)^2=\frac{1152}{64}
 \end{cases}
 \Rightarrow
 \begin{cases}
   x=\frac{-24}{1+2\lambda},\\
   y=\frac{-6}{1+2\lambda},\\
   1+2\lambda=\sqrt{18}
 \end{cases}
 \Rightarrow
 \begin{cases}
   x=\frac{-24}{1+2\lambda},\\
   y=\frac{-6}{1+2\lambda},\\
   \lambda=\frac{\sqrt{18}-1}{2}
 \end{cases}
 \Rightarrow
 \begin{cases}
   x=\frac{-24}{1+\sqrt{18}-1},\\
   y=\frac{-6}{1+\sqrt{18}-1},\\
   \lambda=\frac{\sqrt{18}-1}{2}
 \end{cases}
 \Rightarrow
 \begin{cases}
   x=-\frac{8}{\sqrt{2}},\\
   y=-\frac{2}{\sqrt{2}},\\
   \lambda=\frac{\sqrt{18}-1}{2}
 \end{cases}$$

 Получаем две точки: $(-\frac{8}{\sqrt{2}},-\frac{2}{\sqrt{2}},\frac{\sqrt{18}-1}{2})$ и $(\frac{8}{\sqrt{2}},\frac{2}{\sqrt{2}},-\frac{\sqrt{18}-1}{2})$

Найдем вторые производные:

 $$L^{\prime}_x=4x+12y+2x\lambda\\
   L^{\prime}_y=12x+64y+32y\lambda\\
   L^{\prime}_{\lambda}=x^2+16y^2-64$$

   $$L^{\prime\prime}_{xx}=4+2\lambda, \ \ \ L^{\prime\prime}_{yy}=64+32\lambda, \ \ \ L^{\prime\prime}_{\lambda\lambda}=0$$

   $$L^{\prime\prime}_{xy}=12, \ \ \ L^{\prime\prime}_{x\lambda}=2x, \ \ \ L^{\prime\prime}_{y\lambda}=32y$$

Составим матрицу Гёссе:

   $$\begin{pmatrix}
L''_{\lambda \lambda} & L''_{\lambda x} & L''_{\lambda y}\\ 
L''_{x \lambda} & L''_{xx} & L''_{xy}\\
L''_{y \lambda} & L''_{yx} & L''_{yy}\\

=\begin{pmatrix}
0 & 2x & 32y\\ 
2x & 4+2\lambda & 12\\

Далее найдем определитель в общем виде и разберемся с каждой точкой:

$$\Delta=\begin{vmatrix}
0 & 2x & 32y\\ 
2x & 4+2\lambda & 12\\
32y & 12 & 64+32\lambda\\
\end{vmatrix}=
\begin{vmatrix}
+ & - & +\\

\end{vmatrix}=0\cdot
\begin{vmatrix}

\end{vmatrix}-2x\cdot
\begin{vmatrix}

\begin{vmatrix}
2x & 4+2\lambda\\

\end{vmatrix}=(-2x)\cdot ((2x\cdot(64+32\lambda))-384y)+32y \cdot(24x-(32y\cdot(4+2\lambda)))=(-2x)\cdot(128x+64x\lambda-384y)+32y\cdot(24x-128y-64y\lambda)=-256x^2-128x^2\lambda+768xy+768xy-4096y^2-2048y^2\lambda=-256x^2+1536xy-4096y^2-128x^2\lambda-2048y^2\lambda$$

По условию задачи $x^2+16y^2=64$, тогда

Т.е. знак определителя зависит только от знака $\lambda$.

Если $\lambda=\frac{\sqrt{18}-1}{2}$, то $\Delta<0$, следовательно $(-\frac{8}{\sqrt{2}},-\frac{2}{\sqrt{2}},\frac{\sqrt{18}-1}{2})$ - точка минимума.

**3. Найти производную функции $U=x^2+y^2+z^2$ по направлению вектора** **$\overrightarrow{c}=(-9,8,-12)$ в точке $M (8, -12, 9)$.**

а) Найдем частные производные в точке $M (8, -12, 9)$

$\frac{\vartheta U}{\vartheta x}=2x$ $\ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \frac{\vartheta U}{\vartheta x}\Bigl|_{(8,-12,9)}=2\cdot8=16$

$\frac{\vartheta U}{\vartheta y}=2y$ $\ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \frac{\vartheta U}{\vartheta y}\Bigl|_{(8,-12,9)}=2\cdot(-12)=-24$

$\frac{\vartheta U}{\vartheta z}=2z$ $\ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \frac{\vartheta U}{\vartheta z}\Bigl|_{(8,-12,9)}=2\cdot9=18$

б) Найдем координаты направляющего вектора еденичной длинны

$|\overrightarrow{c}|=\sqrt{x^2_0+y^2_0+z^2_0}=\sqrt{(-9)^2+8^2+(-12)^2}=\sqrt{81+64+144}=\sqrt{289}=17$

$\overrightarrow{c}_0=\frac{\overrightarrow{c}}{|\overrightarrow{c}|}=\Bigl(-\frac{9}{17},\frac{8}{17},-\frac{12}{17}\Bigr)$

в) Вычисляем по формуле

$U^{\prime}_{\overrightarrow{c}}\Bigl|_{(8,-12,9)}=-\frac{9}{17}\cdot16+\frac{8}{17}\cdot(-24)-\frac{12}{17}\cdot18=-\frac{144}{17}-\frac{192}{17}-\frac{216}{17}=-\frac{552}{17}=-32\frac{8}{17}$


**4. Найти производную функции $U=e^{x^2+y^2+z^2}$ по направлению вектора $\overrightarrow{d}=(4,-13,-16)$ в точке $M (-16, 4, -13)$.**

а) Найдем частные производные в точке $M (-16, 4, -13)$

$\frac{\vartheta U}{\vartheta x}=2xe^{x^2+y^2+z^2}$ $\ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \frac{\vartheta U}{\vartheta x}\Bigl|_{(-16, 4, -13)}=-32e^{441}$

$\frac{\vartheta U}{\vartheta y}=2ye^{x^2+y^2+z^2}$ $\ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \frac{\vartheta U}{\vartheta y}\Bigl|_{(-16, 4, -13)}=8e^{441}$

$\frac{\vartheta U}{\vartheta z}=2ze^{x^2+y^2+z^2}$ $\ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \frac{\vartheta U}{\vartheta z}\Bigl|_{(-16, 4, -13)}=-26e^{441}$

б) Найдем координаты направляющего вектора еденичной длинны

$|\overrightarrow{d}|=\sqrt{x^2_0+y^2_0+z^2_0}=\sqrt{4^2+(-13)^2+(-16)^2}=\sqrt{16+169+256}=\sqrt{441}=21$

$\overrightarrow{d}_0=\frac{\overrightarrow{d}}{|\overrightarrow{d}|}=\Bigl(\frac{4}{21},-\frac{13}{21},-\frac{16}{21}\Bigr)$

в) Вычисляем по формуле

$U^{\prime}_{\overrightarrow{d}}\Bigl|_{(-16,4,-13)}=\frac{4}{21}\cdot(-32e^{441})-\frac{13}{21}\cdot8e^{441}-\frac{16}{21}\cdot(-26e^{441})=-\frac{128}{21}e^{441}-\frac{104}{21}e^{441}+\frac{416}{21}e^{441}=e^{441}(-\frac{128}{21}-\frac{104}{21}+\frac{416}{21})=e^{441}(\frac{-128-104+416}{21})=\frac{184}{21}e^{441}=8\frac{16}{21}e^{441}$
