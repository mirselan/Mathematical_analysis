**1. Прочитать высказывания математической логики, построить их отрицания и установить истинность.**

**1.1** 

$\forall y \in [0;1] : sgn(y) = 1$

Для любого у принадлежащего от 0 включительно до 1 включительно верно, что сигнум от у равняется 1.

***Решение:***
$$sgn(y) = \begin{equation*}
 \begin{cases}
   1 &if &y > 0, 
   \\
   0 &if &y = 0,
   \\
   -1 &if &y < 0.
 \end{cases}
\end{equation*}$$

Высказывание является ложным по определению сигнум для 0.

Построим отрицание данного высказывания: $\exists y \in [0; 1] : sng(y) \neq 1$

Существует у из промежутка от 0 включительно до 1 включительно, что сигнум от у не равен 1. Утверждение истинно, так как $sgn(0) \neq 1$, значит текущее высказывание ложное.

***Ответ:*** Текущее высказывание ложное.

**1.2**

$\forall n \in \mathbb {N} > 2 : \exists x, y, z \in \mathbb {N} : x^n = y^n + z^n$ 

Для любого n, принадлежащего множеству натуральных чисел и большему чем 2, верно, что существуют такие x, y и z, являющиеся натуральными числами, для которых верно, что х в степени n равен сумме у в степени n и z в степени n.

***Решение:***

Теорема Ферма утверждает, что для любого натурального числа n > 2 уравнение:
$$a^n + b^n = c^n$$
не имеет решений в целых ненулевых числах a , b , c.

Исходя из теоремы Ферма, высказывание является ложным.

Построим отрицание данного высказывания: $\exists n \in \mathbb {N} > 2 : \forall x, y, z \in \mathbb {N} : x^n \neq y^n + z^n$ 

Существует n, принадлежащее натуральным числам и большее чем 2, такое, что для любых х, у и z, являющихся натуральными числами, верно, что х в степени n не равно сумме у в степени n и z в степени n. Утверждение истинно исходя из теоремы Ферма, значит текущее высказывание ложно.

***Ответ:*** Текущее высказывание ложно.

**1.3**

$\forall x \in \mathbb {R}\exists X \in \mathbb {R} : X > x$

Для любого х, принадлежащего к множеству вещественных чисел, существует Х, также принадлежащее множеству вещественных чисел такое, что Х больше чем х.

***Решение:***

Х и х принадлежат множеству вещественных чисел, соответственно для любого х найдется Х большее чем х.Исходя из этого рассуждения высказывание является истинным.

Построим отрицание данного высказывания: $\exists x \in \mathbb {R}\forall X \in \mathbb {R} : X < x$

Существует х, принадлежащее к множеству вещественных чисел, такое, что каждое Х, также принадлежащее множеству вещественных чисел будет меньше чем х. Утверждение ложно так как множество вещественных чисел бесконечно, значит текущее высказывание истинно.

***Ответ:*** Текущее высказывание является истинным.


**1.4**

$\forall x \in \mathbb {C}\nexists y \in \mathbb {C} : x > y||x < y$

Для любого х, принадлежащего множеству комплексных чисел, не существует у, также принадлежащий множеству комплексных чисел, чтобы было верно, что х больше у или х меньше у.

***Решение:***
Как известно из свойств комплексных чисел, нельзя указать, какое из двух комплексных чисел больше или меньше. Отсюда следует, что высказывание истинно.

Построим отрицание данного высказывания: $\exists x \in \mathbb {C}\exists y \in \mathbb {C} : x < y||x > y$


Существует х, принадлежащий множеству комплексных чисел, для которого существует у, также принадлежащий множеству комплексных чисел, такой, что х меньше у или х больше у. Утверждение ложно так как комплексные числа не могут быть больше или меньше друг друга, значит текущее высказывание истинно.

***Ответ:***

Текущее высказывание является истинным.

**1.5**

$\forall y \in [0; \frac{\pi}{2}]\exists\varepsilon > 0 : sin y < sin(y + \varepsilon)$

Для любого у, принадлежащего от 0 включительно до $\frac{\pi}{2}$ включительно, и хотя бы одного $\varepsilon$ больше нуля верно, что синус у меньше чем синус суммы у и $\varepsilon$.

***Решение:***

Синус у имеет максимальное значение равное 1 для у=$\frac{\pi}{2}$, дальнейшее увеличение у не приводит к повышению значения синус у. Отсюда следует, что данное высказывание ложно.

Построим отрицание данного высказывания: $\exists y \in [0; \frac{\pi}{2}]\forall\varepsilon > 0 : sin y > sin(y + \varepsilon)$

Существует хотя бы один у, принадлежащий интервалу от 0 включительно до $\frac{\pi}{2}$ включительно, для которого с каждым $\varepsilon$ больше нуля верно, что синус у больше чем синус суммы у и $\varepsilon$. $sin\frac{\pi}{2}$ = 1, это максимальное значение синуса, отсюда следует что утверждение истинно. Значит текущее высказывание ложно.

***Ответ:*** Текущее высказывание ложно.



**1.6**

$\forall y \in [0; \pi)\exists\varepsilon > 0 : cos y > cos(y + \varepsilon)$

Для любого у, принадлежащего от 0 включительно до $\pi$ не включительно, и хотя бы одного $\varepsilon$, большего чем 0, верно, что косинус у больше чем косинус суммы у и $\varepsilon$.

***Решение:***

Косинус у имеет наибольшее значение, равное 1, при у=0. При увеличении у значение косинуса у уменьшается, следовательно данное высказывание истинно.

Построим отрицание данного высказывания: $\exists y \in [0; \pi)\forall\varepsilon > 0 : cos y < cos(y + \varepsilon)$

Существует такой у принадлежащий от 0 включительно до $\pi$ не включительно, для которого с каждым $\varepsilon$, большим чем 0, верно, что косинус у меньше чем косинус суммы у и $\varepsilon$. Исходя из того, что максимальное значение косинус у имеет при у=0, утверждение ложно, значит текущее высказыванеи истинно.

***Ответ:*** Текущее высказывание истинно.

**1.7**

$\exists x : x \notin \{\mathbb {N}, \mathbb {Z}, \mathbb {Q}, \mathbb {R}, \mathbb {C}\}$

Существует х для которого верно, что х не принадлежит ни к натуральным числам, ни к целым числам, ни к рациональным числам, ни к вещественным числам, ни к комплексным числам.

Построим отрицание данного высказывания: $\forall x : x \in \{\mathbb {N}, \mathbb {Z}, \mathbb {Q}, \mathbb {R}, \mathbb {C}\}$

Для любого х верно, что х принадлежит либо к натуральным числам, либо к целым числам, либо к рациональным числам, либо к вещественным числам, либо к комплексным числам. Утверждение истинно так как множество $\{\mathbb {N}, \mathbb {Z}, \mathbb {Q}, \mathbb {R}, \mathbb {C}\}$ охватывает весь спектр известных чисел, значит текущее высказывание ложно.

***Ответ:*** Текущее высказывание ложно.
